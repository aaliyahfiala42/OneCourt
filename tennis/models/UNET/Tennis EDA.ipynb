```python
import csv
import json
 
# Function to convert a CSV to JSON
# Takes the file paths as arguments
def make_json(csvFilePath, jsonFilePath):
     
    # create a dictionary
    data = {}
     
    # Open a csv reader called DictReader
    with open(csvFilePath, encoding='utf-8') as csvf:
        csvReader = csv.DictReader(csvf)
         
        # Convert each row into a dictionary
        # and add it to data
        for rows in csvReader:
            # Assuming a column named 'No' to
            # be the primary key
            key = rows['positions']
            data[key] = rows
 
    # Open a json writer, and use the json.dumps()
    # function to dump data
    with open(jsonFilePath, 'w', encoding='utf-8') as jsonf:
        jsonf.write(json.dumps(data, indent=4))
         
```

## Evaluate Model: Recall, Precsion, F1

```python
actual = label[0,0] * 255
predict = output[0,0] * 255
```

```python
def precision_recall_f1(actual, predict, thresh): 
    #set # pixels to include in threshhold each direction
    
    tp = 0
    fp = 0
    fn = 0
    tn = 0

    for x in range(actual.shape[0]):
        for y in range(actual.shape[1]):

            #flags
            actual_true = 0
            guess_true = 0
            
            if x % thresh == 0 and y % thresh == 0: 
                #get actuals threshhold box
                for x2 in range(max(x - thresh, 0), min(x + thresh, actual.shape[0])):
                    for y2 in range(max(y - thresh, 0), min(y + thresh, actual.shape[1])):
                        #if tennis ball in threshold
                        if actual[x2, y2] > 10:
                            actual_true = 1

                        #check if guess is in threshhold: 
                        if predict[x2, y2] > 10:
                            guess_true = 1

                #true positive
                if actual_true == 1 and guess_true == 1:
                    tp += 1
                #false positive
                if actual_true == 0 and guess_true == 1:
                    fp += 1
                #false negative
                if actual_true == 1 and guess_true == 0:
                    fn += 1
                #true negative
                if actual_true == 0 and guess_true == 0:
                    tn += 1

    if tp + fp == 0:
        precision = 0.0
    else:
        precision = tp / (tp + fp)
        
    if tp + fn == 0: 
        recall = 0.0
    else: 
        recall = tp / (tp + fn)
    
    if precision + recall == 0:
        f1 = 0.0
    else:
        f1 = (2 * precision * recall) / (precision + recall)
                
    return precision, recall, f1
```

```python
prec_ = []
rec_ = []
f1_ = []
for i, (data, label) in enumerate(test_loader):
    output = model(data[:, 0:3], data[:, 3:6], data[:, 6:9])
    predict = output[0,0] * 255

    actual = label[0,0] * 255

    p, r, f = precision_recall_f1(actual, predict, 100)
    
    prec_.append(p)
    rec_.append(r)
    f1_.append(f)
    
    if i % 5 == 0:
        print(i, ", f1 score: ", f1_[-1])
    
    if i == 10:
        break
```

```python
print("UNET 3 Frame")
print("Precision: ", sum(prec_)/len(prec_))
print("Recall: ", sum(rec_)/len(rec_))
print("F1 Score: ", sum(f1_)/len(f1_))
```
